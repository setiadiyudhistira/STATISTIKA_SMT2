```python
#array 2D
array_2d = ([10,20,30], 
            [40,50,60], 
            [70,80,90])

array_2d = np.append(array_2d, [[100,110,120]], axis=0)
print(array_2d)

array_2d = np.append(array_2d, 
                    [[130, 140, 150]], axis=0)
print(array_2d)
```

```python
#array 2D
array_2d = ([10,20,30], 
            [40,50,60], 
            [70,80,90])

array_2d = np.append(array_2d, [[100,110,120]], axis=0)
print(array_2d)

array_2d = np.append(array_2d, 
                    [[130, 140, 150]], axis=0)
print(array_2d)

array_2d = np.delete(array_2d, 0, axis=0)
print(array_2d)
array_2d = np.delete(array_2d, 1, axis=0)
print(array_2d)

maximun = np.max(array_2d)
minimun = np.min(array_2d)

print(maximun)
print(minimun)

mengurutkan =np.sort(array_2d, axis=0)
print(mengurutkan)
```
