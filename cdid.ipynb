class pertama

```python
class Kendaraan:
    def __init__(self, merk, model):
        self.merk = merk
        self.model = model

    def info(self):
        return f"{self.merk} {self.model}"

mobil = Kendaraan ("Kia", "Picanto") 
print(mobil.info())
```

class kedua

```python
class Kendaraan:
    def __init__(self, merek, model, harga):
        self.merek = merek 
        self.model = model
        self.harga = harga
    
    def info(self):
        return f"{self.merek} {self.model} {self.harga}"
    
    def info_harga(self):
        output = f"Harga kendaraan dengan merek {self.merek} adalah {self.harga}"
        return output
    
mobil = Kendaraan("Nissan", "Slyphy", "Rp 53.293.302")
print(mobil.info_harga())
```

class tersendiri

```python
class Mahasiswa:
    def __init__(self, nama, nim, kelas):
        self.nama = nama
        self.nim = nim
        self.kelas = kelas

    def perkenalan(self):
        print(f"perkenalkan nama saya {self.nama} dengan nim {self.nim} dari kelas {self.kelas}")

identitas = Mahasiswa("Satya", "25090000", "1-A") 
print(identitas.perkenalan())
```

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def enqueue(self,item):
        item = self.queue.append(item)
        print(f"{item} ditambahkan ke dalam queue.")

    def dequeue(self):
        item = self.queue.pop(0)
        print(f"{item} dihapus dari queue.")
        
    def display(self):
        return self.queue
    
queue = Queue
queue.enqueue(10)
queue.enqueue(20) 

queue.enqueue
queue.dequeue
queue.display
```
